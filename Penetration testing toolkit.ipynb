```python
import socket
import requests
import threading


# MODULE 1: PORT SCANNER

def port_scanner(target, ports):
    print(f"\n[+] Scanning Target: {target}")
    print("[+] Scanning Ports...\n")

    for port in ports:
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        s.settimeout(0.5)
        result = s.connect_ex((target, port))

        if result == 0:
            print(f"PORT {port} : OPEN")
        s.close()



# MODULE 2: DIRECTORY BRUTE-FORCER

def dir_bruteforce(url, wordlist):
    print(f"\n[+] Running Directory Bruteforce on {url}\n")
    for word in wordlist:
        full_url = url + "/" + word
        try:
            r = requests.get(full_url)
            if r.status_code == 200:
                print(f"[FOUND] {full_url}")
            elif r.status_code == 403:
                print(f"[FORBIDDEN] {full_url}")
        except:
            pass



# MODULE 3: SUBDOMAIN FINDER

def subdomain_scanner(domain, subdomains):
    print(f"\n[+] Scanning for Subdomains of {domain}\n")
    for sub in subdomains:
        url = f"http://{sub}.{domain}"
        try:
            requests.get(url)
            print(f"[ACTIVE] {url}")
        except:
            pass



# MAIN TOOLKIT MENU

def main():
    while True:
        print("\n=== PENETRATION TESTING TOOLKIT ===")
        print("1. Port Scanner")
        print("2. Directory Brute-Forcer")
        print("3. Subdomain Scanner")
        print("4. Exit")

        choice = input("\nSelect a module: ")

        if choice == "1":
            target = input("Enter target IP/domain: ")
            ports = [21, 22, 23, 25, 53, 80, 110, 443, 3306, 8080]
            port_scanner(target, ports)

        elif choice == "2":
            url = input("Enter website URL (e.g., https://example.com): ")
            wordlist = ["admin", "login", "test", "dashboard", "uploads", "images"]
            dir_bruteforce(url, wordlist)

        elif choice == "3":
            domain = input("Enter domain name (e.g., google.com): ")
            common_subs = ["www", "mail", "test", "dev", "ftp", "api"]
            subdomain_scanner(domain, common_subs)

        elif choice == "4":
            print("Exiting Toolkit...")
            break

        else:
            print("Invalid choice, try again!")


if __name__ == "__main__":
    main()
```
